# Spotify Music Project

### Dataset cleaning and saving

```python
# Reading all Music Genre Datasets
alt = pd.read_csv('alternative_music_data.csv') # Alternative-Rock Music Dataset
blues = pd.read_csv('blues_music_data.csv') # Blues Music Dataset
hhop = pd.read_csv('hiphop_music_data.csv') # Hip-Hop Music Dataset
indie = pd.read_csv('indie_alt_music_data.csv') # Indie-Rock Music Dataset
metal = pd.read_csv('metal_music_data.csv') # Metal Music Dataset
pop = pd.read_csv('pop_music_data.csv') # Pop Music Dataset
rock = pd.read_csv('rock_music_data.csv') # Rock Music Dataset
```

```python
#lets see if there are any columns with missing values 
alt_columns=alt.columns[alt.isnull().any()]
alt[alt_columns].isnull().sum()
```

```python
blues_columns=blues.columns[blues.isnull().any()]
blues[blues_columns].isnull().sum()
```

```python
hhop_columns=hhop.columns[hhop.isnull().any()]
hhop[hhop_columns].isnull().sum()
```

```python
indie_columns=indie.columns[indie.isnull().any()]
indie[indie_columns].isnull().sum()
```

```python
metal_columns=metal.columns[metal.isnull().any()]
metal[metal_columns].isnull().sum()
```

```python
pop_columns=pop.columns[pop.isnull().any()]
pop[pop_columns].isnull().sum()
```

```python
rock_columns=rock.columns[rock.isnull().any()]
rock[rock_columns].isnull().sum()
```

```python
# Dropping unnecessary columns
alt.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
blues.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
hhop.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
indie.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
metal.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
pop.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
rock.drop(['track_href','uri', 'analysis_url'], axis=1, inplace=True)
```

```python
# Saving Cleaned Files for Exploratory Data Analysis (EDA)
alt.to_csv("altdata_cleansed.csv", encoding="utf-8")
blues.to_csv("bluesdata_cleansed.csv", encoding="utf-8")
hhop.to_csv("hhopdata_cleansed.csv", encoding="utf-8")
indie.to_csv("indiedata_cleansed.csv", encoding="utf-8")
metal.to_csv("metaldata_cleansed.csv", encoding="utf-8")
pop.to_csv("popdata_cleansed.csv", encoding="utf-8")
rock.to_csv("rockdata_cleansed.csv", encoding="utf-8")
```
